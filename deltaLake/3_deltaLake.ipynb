### **SCHEMA ENFORCEMENT**

```python
data = [(1,23434,"harish",2),(2,33434,"thark",3),(2,34535,"mad",4)]
df = spark.createDataFrame(data,["id","salary","name","experience"])

display(df)
```

```python
df.write.format("delta")\
    .mode("append")\
    .save("/Volumes/deltalake/default/deltavol")
```

### **SCHEMA EVOLUTION**

```python
df.write.format("delta")\
    .mode("append")\
    .option("mergeSchema",True)\
    .save("/Volumes/deltalake/default/deltavol/sink/")
```

### **READ DELTA DATA**

```python
%sql
SELECT * from delta.`/Volumes/deltalake/default/deltavol/sink/`
```

### **SCHEMA OVERWRITE**

```python
data = [(1,28000,"firstSalary","Accenture"),(2,32000,"secondSalary","Accenture"),(3,42000,"thirdSalary","Accenture")]

df1 = spark.createDataFrame(data,["id","salary","stagesOfSalary","Company"])

display(df1)
```

```python
df1.write.format("delta")\
    .mode("overwrite")\
    .option("overwriteSchema",True)\
    .save("/Volumes/deltalake/default/deltavol/sink/")
```
