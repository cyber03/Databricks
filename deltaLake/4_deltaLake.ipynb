### **DML**

```python
data = [(5,23232,"hari",1),(6,34342,"Harish",2),(7,343332,"Harishbabu",3)]
df = spark.createDataFrame(data,["id","CTC","name","experience"])
display(df)
```

```python
df.write.format("delta")\
    .mode("append")\
    .save("/Volumes/deltalake/default/deltavol/sinkDML/")
```
